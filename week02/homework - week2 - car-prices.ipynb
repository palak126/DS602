## Question 4. Analyze the relationship between the wholeprice `mmr` and the `salesprice`.
- What is the overall correlation?  
- What happens to the correlation as you increase the maximum selling price window (e.g., 0-10000, 0-20000,...)?
- Create a scatter plot with the relationship.  
- Create a graph that shows how the correlation changes.

````python
corr = df['mmr'].corr(df['sellingprice'])
print(f"Overall correlation : {corr}")
````

this correlation coefficient explains that mmr and selling price are directly correlated to each other if mmr increases, selling price increases or mmr = 0.9835 * sellingprice

````python
# defined the selling price window to check correlation change

max_prices = [10000, 20000, 30000, 40000, 50000]
corr = []

for max_price in max_prices:
    filtered_df = df[df['sellingprice'] <= max_price]
    correlation = filtered_df['mmr'].corr(filtered_df['sellingprice'])
    corr.append(correlation)
````

````python
plt.figure(figsize=(10, 6))
plt.scatter(df['mmr'], df['sellingprice'], alpha=0.5)
plt.title('Relationship between MMR and Selling Price')
plt.xlabel('MMR')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()
````

This graph represents a direct relation between mmr and selling price. If mmr increases, the selling price also increases

````python
plt.figure(figsize=(10, 6))
plt.plot(max_prices, corr, marker='o')
plt.title('Correlation vs. Maximum Selling Price Window')
plt.xlabel('Maximum Selling Price Window')
plt.ylabel('Correlation Coefficient')
plt.grid(True)
plt.xticks(max_prices)
plt.show()
````

The correlation coefficient grows for the max selling price window from 10000-20000 most rapidly whereas grows less for other max selling windows further

## Question 5. How does the distribution of `salesprice` vary by make for cars made between 2010-2015?
- Use the 10 most common makes (your cleaned version of this column).  
- Create a graph that answers this question.

````python
# filtering df for year 2010-2015 and taking top 10 make values

filtered_df = df[(df['year'] >= 2010) & (df['year'] <= 2015)]
top_10_makes = filtered_df['make'].value_counts().head(10).index.tolist()

plt.figure(figsize=(12,8))
for make in top_10_makes:
    make_data = filtered_df[filtered_df['make'] == make]
    plt.hist(make_data['sellingprice'], bins=20, label=make)
    
plt.title('distribution of selling price by make(2010-2015)')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
````
