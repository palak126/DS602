## Question 1. We need to know what's included in this data.
- How many countries are present?  
- What's the timespan of the data?  
- Does each country have a row for every year present?  
- Which countries are missing years?  
- How many rows have at least 1 non-null values across the columns 3-57?

````python
# printing the number and names of distinct countries in the dataset
countries_count = df['country'].nunique()
distinct_countries = df['country'].unique()

print("Number of distinct countries:", countries_count)
print("Distinct countries:", distinct_countries)
````

````python
# calculating timespan by subtracting max year by min year
min_year = df['year'].min()
max_year = df['year'].max()

timespan = max_year - min_year

print('Data Timespan:', timespan)
````

````python
# calculating if number of unique years for each country is equal to the total number of unique years
grouped_data = df.groupby('country')['year'].nunique()
each_country_year = all(grouped_data == df['year'].nunique())

print("Does each country have a row for every year present?", each_country_year)
````

````python
# printing countries that have missing years
missing_years = df.groupby('country')['year'].apply(lambda x: set(range(df['year'].min(), df['year'].max() + 1)) - set(x))

countries_with_missing_years = missing_years[missing_years.apply(lambda x: len(x) > 0)]

print("Countries with missing years:")
print(countries_with_missing_years)
````

````python
# printing rows that have at least 1 non-null values across the columns 3-57
rows_with_non_null = df.iloc[:, 3:58].notnull().any(axis=1)

count_rows_with_non_null = sum(rows_with_non_null)

print("Number of rows with at least one non-null value across columns 3-57:", count_rows_with_non_null)
````

## Question 2: Wide data isn't usually helpful.
Convert this to a long format with the following columns:  
1. country  
2. year  
3. diagnosis method: mapping should be: 

```
  {'newsp':'Pulmonary Smear (+)'
  ,'newsn':'Pulmonary Smear (-)'
  ,'newep':'Extrapulmonary'
  ,'newrel':'Relapse'}   
 ```

4. gender: male or female.   
5. age: lower_age - higher_age, e.g., 0-14  
6. number of cases  

__Note__: You can use `replace` (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)

````python
df.sample(5)
````

````python
# converting wide format dataframe into long format dataframe using pd.melt
dfLong = (
    pd.melt(df, id_vars=['country','year'], value_vars=['new_sp_m014', 'new_sp_m1524',
       'new_sp_m2534', 'new_sp_m3544', 'new_sp_m4554', 'new_sp_m5564',
       'new_sp_m65', 'new_sp_f014', 'new_sp_f1524', 'new_sp_f2534',
       'new_sp_f3544', 'new_sp_f4554', 'new_sp_f5564', 'new_sp_f65',
       'new_sn_m014', 'new_sn_m1524', 'new_sn_m2534', 'new_sn_m3544',
       'new_sn_m4554', 'new_sn_m5564', 'new_sn_m65', 'new_sn_f014',
       'new_sn_f1524', 'new_sn_f2534', 'new_sn_f3544', 'new_sn_f4554',
       'new_sn_f5564', 'new_sn_f65', 'new_ep_m014', 'new_ep_m1524',
       'new_ep_m2534', 'new_ep_m3544', 'new_ep_m4554', 'new_ep_m5564',
       'new_ep_m65', 'new_ep_f014', 'new_ep_f1524', 'new_ep_f2534',
       'new_ep_f3544', 'new_ep_f4554', 'new_ep_f5564', 'new_ep_f65',
       'newrel_m014', 'newrel_m1524', 'newrel_m2534', 'newrel_m3544',
       'newrel_m4554', 'newrel_m5564', 'newrel_m65', 'newrel_f014',
       'newrel_f1524', 'newrel_f2534', 'newrel_f3544', 'newrel_f4554',
       'newrel_f5564', 'newrel_f65'], var_name= 'diagnosis', value_name='total_cases')
)

dfLong.sample(10)
````

````python
# extracting diagnosis method type from the column diagnosis
dfLong['diagnosis_type'] = dfLong['diagnosis'].str.extract(r'(new_sp|new_ep|newrel|new_sn)')
dfLong.sample(10)
````

````python
# replacing the diagnosis type by space so that column be left with gender and age 
dfLong['diagnosis'] = dfLong['diagnosis'].str.replace('new_sp_', '')
dfLong['diagnosis'] = dfLong['diagnosis'].str.replace('new_sn_', '')
dfLong['diagnosis'] = dfLong['diagnosis'].str.replace('newrel_', '')
dfLong['diagnosis'] = dfLong['diagnosis'].str.replace('new_ep_', '')
dfLong.sample(20)
````

````python
# extracting gender from the diagnosis column using lambda
dfLong['gender'] = dfLong['diagnosis'].apply(lambda x: 'male' if x.startswith('m') else ('female' if x.startswith('f') else None))
dfLong.sample(10)
````

````python
# assigned a dictionary for values that were not in a proper format
replacement_dict = {
    'm014': 'm0014',
    'm65': 'm6580',
    'f65': 'f6580',
    'f014': 'f0014'
}

# replacing the values in the column based on the dictionary and extracting it in a new column
dfLong['diagnosis'] = dfLong['diagnosis'].replace(replacement_dict)
dfLong['age'] = dfLong['diagnosis'].str.extract(r'(\d{2})(\d{2})').apply(lambda x: f"{x[0]}-{x[1]}", axis=1)
dfLong.sample(10)
````

````python
dfLong.drop(columns='diagnosis', inplace=True)
````

## Question 3: Create a graph that shows when countries started to report TB cases.

````python
# changing nan values in total_cases column to 0 because there should be a value in total_cases column when the countries start to report TB cases
dfLong['total_cases'].fillna(0, inplace=True)
dfLong.sample(10)
````

````python
# plotting graph of total number of countries vs year when they started reporting TB cases
start_year = dfLong.groupby('country')['year'].min()

plt.figure(figsize=(10,6))
start_year.value_counts().sort_index().plot(kind='line', marker='o')
    
plt.title('When countries started to report TB cases')
plt.xlabel('Year')
plt.ylabel('Total number of Countries')
plt.grid(True)
plt.show()
````
